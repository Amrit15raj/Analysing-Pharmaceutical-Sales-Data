```python
#On which day of the week is the second drug (M01AE) most often sold?
import pandas as pd

df = pd.read_csv("data/salesdaily.csv")

weekday_sales = df.groupby("Weekday Name")["M01AE"].sum()

res = weekday_sales.idxmax()
volume = weekday_sales.max()

print(f"M01AE is most often sold on {res} with a volume of {volume:.2f}.")
```

```python
#Which three drugs have the highest sales in January 2015, July 2016, September 2017

# Convert date column to datetime
df["Datum"] = pd.to_datetime(df["datum"])

# Drug columns
drug_cols = ["M01AB", "M01AE", "N02BA", "N02BE", "N05B", "N05C", "R03", "R06"]

months = [
    ("January 2015", "2015-01"),
    ("July 2016", "2016-07"),
    ("September 2017", "2017-09")
]

for month_name, period in months:
    month_df = df[df["Datum"].dt.to_period("M") == period]
    
    top3 = month_df[drug_cols].sum().sort_values(ascending=False).head(3)
    print(f"\nTop 3 drugs in {month_name}:")
    print(top3)
```

```python
#Which drug has sold most often on Mondays in 2017?

df_2017 = df[df["Datum"].dt.year == 2017]
week_sales = df_2017.groupby("Weekday Name")[drug_cols].sum()
monday_sales = week_sales.loc["Monday"].sort_values(ascending=False)
print(f"The drug most often sold on Mondays in 2017 is {monday_sales.idxmax()} with the volume of {monday_sales.max():.2f}")
```

```python
#What medicine sales may be in January 2020?
import numpy as np
from sklearn.linear_model import LinearRegression

# Remove rows containing 0 values
df = df[(df[drug_cols] != 0).all(axis=1)]

# Remove incomplete October 2019
df = df[df["Datum"] < "2019-10-01"]

# Monthly sales of N02BA
monthly_sales = df.resample("M", on="Datum")["N02BA"].sum()

#print(monthly_sales.to_string())

# Month numbers
X = np.arange(len(monthly_sales)).reshape(-1, 1)

# Sales values
y = monthly_sales.values

# Train model
model = LinearRegression()
model.fit(X, y)

# January 2020 (4 months after September 2019)
january_2020_index = len(monthly_sales) + 4

prediction = model.predict([[january_2020_index]])

print(f"Predicted N02BA sales for January 2020: {prediction[0]:.2f}")
```
